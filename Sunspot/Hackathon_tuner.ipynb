```python
import h2o
from h2o.automl import H2OAutoML
```

```python
h2o.init()
```

```python
data = h2o.import_file('train_.csv')
```

```python
features=data.columns
```

```python
features
```

```python
output='Avg_sunspot_count'
```

```python
aml= H2OAutoML(max_models=7, max_runtime_secs=600, seed=2)
```

```python
aml.train(x = features, y = output, training_frame = data)
```

```python
lb=aml.leaderboard
lb.head()
```
